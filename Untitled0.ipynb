```python
import pandas as pd

# upload file
from google.colab import files
uploaded = files.upload()

# read csv
df = pd.read_csv(list(uploaded.keys())[0])

df.head()
```

```python
df.isnull().sum()
```

```python
rating_columns = df.columns[2:]  # Sr no. & Student ID hata kar

avg_ratings = df[rating_columns].mean()
avg_ratings
```

```python
import matplotlib.pyplot as plt

avg_ratings.sort_values().plot(kind='barh', figsize=(8,5))
plt.title("Average Rating by Course Feedback Parameters")
plt.xlabel("Average Rating")
plt.show()
```
